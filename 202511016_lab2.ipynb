```python
def generate_data(n, noise=0.0, seed=None):
    if seed is not None:
        np.random.seed(seed)
    X = np.random.randn(n)
    Y = (X > 0).astype(int)
    # Flip labels with probability noise
    flips = np.random.rand(n) < noise
    Y[flips] = 1 - Y[flips]
    return X, Y
```
